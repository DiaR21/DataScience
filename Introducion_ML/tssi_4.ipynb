```python
#Selección univariada
#Class sklearn.feature_selection SelectKBest(score_func=<function f_classif>, k=10)
#f_classif: ANOVA F-valor entre etiqueta/parámetro para tareas de clasificación 
#mutual_info_classif: Información mutua para un objetivo discreto
#chi2: Estadístico Chi-cuadrado para parámetros no negativos para tareas de clasificación
#f_regression: F-valor entre etiqueta/parámetro para tareas de regresión 
#mutual_info_regression: Información mutua para un objetivo continuo
#SelectPercentile: Selecciona parámetros basados en el percentil de los puntajes más altos 
#SelectFpr: Selecciona parámtros basado en la prueba de tasa de falsos positivos
#SelectFdr: Selecciona parámetros basado en una tasa estimada de false discovery rate
#SelectFwe: Selecciona parámetros basado en una family-wise rate
#GenericUnivariateSelect: Selector de parámetro univariado con modo configurable
```

```python
# Seleccion unitaria
from pandas import read_csv
from numpy import  set_printoptions
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
filename='diabetes.csv'
names=['preg', 'plas', 'pres', 'skin', 'test', 'mass','pedi', 'age','class']
dataframe=read_csv(filename, names=names)
array=dataframe.values
X=array[:,0:8]
Y=array[:,8]
test=SelectKBest(score_func=chi2, k=4)
fit=test.fit(X,Y)
set_printoptions(precision=3)
print(fit.scores_)
feactures=fit.transform(X)
print(feactures[0:6,:])
```

```python
# Eliminscion recursiva de parametros
from pandas import read_csv
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
filename='diabetes.csv'
names=['preg', 'plas', 'pres', 'skin', 'test', 'mass','pedi', 'age','class']
dataframe=read_csv(filename, names=names)
array=dataframe.values
X=array[:,0:8]
Y=array[:,8]
model= LogisticRegression()
rfe=RFE(model,5)
fit=rfe.fit(X,Y)
print("Num Parametros: %d"%fit.n_features_)
print("Parametros seleccionados: %s"%fit.support_)
print("Ranking de parametros: %s"%fit.ranking_)
```

```python
#Analisis de componentes principales
from pandas import read_csv
from sklearn.decomposition import PCA

filename='diabetes.csv'
names=['preg', 'plas', 'pres', 'skin', 'test', 'mass','pedi', 'age','class']
dataframe=read_csv(filename, names=names)
array=dataframe.values
X=array[:,0:8]
Y=array[:,8]

pca=PCA(n_components=3)
fit=pca.fit(X)
print("Varianza:\n %s"%fit.explained_variance_ratio_)
print(fit.components_)
```
