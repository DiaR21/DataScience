```python
# simetria, ajuste de bondad de distribucion de probabilidad gaussiana
#funcion  skew()
#asimetría = 0: distribuido normalmente.
#asimetría> 0: más peso en la cola izquierda de la distribución.
#asimetría <0: más peso en la cola derecha de la distribución. 
from pandas import read_csv
filename='diabetes.csv'
names=['preg', 'plas', 'pres', 'skin', 'test', 'mass','pedi', 'age','class']
data=read_csv(filename, names=names)
print(data.skew()) # skewness=(3(mean-median)/Desv)
```

```python
# Graficos univariados: Graficas de matriz de correlacion
from pandas import read_csv
from matplotlib import pyplot as plt
import numpy
filename='diabetes.csv'
names=['preg', 'plas', 'pres', 'skin', 'test', 'mass','pedi', 'age','class']
data=read_csv(filename, names=names)
correlations=data.corr()
fig=plt.figure()
ax=fig.add_subplot(111)  #"111" significa "cuadrícula 1x1, primer subparcela" y 
                          #"234" significa "cuadrícula 2x3, cuarta parcela"
                          
cax=ax.matshow(correlations, vmin=-1, vmax=1)
fig.colorbar(cax)
plt.show()
```

```python
from pandas import read_csv
from matplotlib import pyplot as plt
import numpy
filename='iris_01.csv'
names=['sepal_length', 'selpal_width', 'petal_length', 'petal_width','class']
data=read_csv(filename, names=names)
correlations=data.corr()
fig=plt.figure()
ax=fig.add_subplot(111)  #"111" significa "cuadrícula 1x1, primer subparcela" y 
                          #"234" significa "cuadrícula 2x3, cuarta parcela"
                          
cax=ax.matshow(correlations, vmin=-1, vmax=1)
fig.colorbar(cax)
ticks=numpy.arange(0,5,1) # para los ubicaciones de las etiquetas 
ax.set_xticks(ticks) #Establezca las ubicaciones de las marcas de xaxis.
ax.set_yticks(ticks)
ax.set_xticklabels(names)
ax.set_yticklabels(names)
plt.show()
```
